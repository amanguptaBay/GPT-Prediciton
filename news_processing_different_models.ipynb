# Reading Data

# Run Sample through GPT-3.5 (Amazon Paper Replication)

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_community.llms import Ollama
import langchain
```

```python
gpt = ChatOpenAI(model="gpt-3.5-turbo")
mistral = Ollama(model="mistral")
ollama = Ollama(model="tinyllama")
```

```python
news_summary_prompt = ChatPromptTemplate.from_template("Summarize the following noisy news data and identify keywords of the news relevant to a Apple stock analyst. Provide a formatted answer as such Summary: ..., Keywords: ... News: {news}")
news_summary_chain = news_summary_prompt | gpt | StrOutputParser()
gpt_results = news_summary_chain.invoke({"news": sampled_news.iloc[-1]["fetched_body"]})
print(gpt_results)
```

```python
from langchain.chains.combine_documents.refine import RefineDocumentsChain
from langchain_core.prompts import PromptTemplate
from langchain.chains import (
    LLMChain,
)
document_prompt = PromptTemplate(
    input_variables=["page_content"],
     template="{page_content}"
)
document_variable_name = "context"
initial_prompt = PromptTemplate.from_template(
    "Summarize the following noisy news data and identify news relevant to an Apple stock analyst. News: {context}"
)
initial_chain = LLMChain(llm=ollama, prompt=initial_prompt)
initial_response_name = "prev_response"
reduce_prompt = PromptTemplate.from_template(
    "Old summary: {prev_response}. "
    "Combine old summary with this additional information: {context}"
)
reduce_llm_chain = LLMChain(llm=ollama, prompt=reduce_prompt)

chain = RefineDocumentsChain(
    initial_llm_chain=initial_chain,
    refine_llm_chain=reduce_llm_chain,
    document_prompt=document_prompt,
    document_variable_name=document_variable_name,
    initial_response_name=initial_response_name,
)
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator=".",
    chunk_size=200,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)
```

```python
text_splitter.create_documents([sampled_news.iloc[-1]["fetched_body"],])
```

```python
out = chain.invoke(text_splitter.create_documents([sampled_news.iloc[-1]["fetched_body"],]))
out
```
