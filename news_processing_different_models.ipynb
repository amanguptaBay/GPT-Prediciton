# Reading Data
Read in news data, drop duplicates then parse published date into Timestamp object

```python
news = pd.read_csv(".data/organized_news.csv")
news.drop([news.columns[0]], axis=1, inplace=True)
news.head()
```

```python
news.drop_duplicates(["url"])
```

```python
import datetime
import dateutil.parser
```

```python
news['published date'] = news['published date'].apply(lambda x: dateutil.parser.parse(x))
```

```python
news['week'] = news['published date'].apply(lambda x: x.week)
news['year'] = news['published date'].apply(lambda x: x.year)
```

# Sampling
Too much data, for each week, year only selecting 2 articles to process

```python
news
```

```python
sampled_news = news.groupby(["year","week"]).sample(2, random_state=1)
```

```python
sampled_news
```

```python
sampled_news["fetched_body"].apply(lambda x: x.upper())
```

# Run Sample through GPT-3.5 (Amazon Paper Replication)

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_community.llms import Ollama
import langchain
```

```python
gpt = ChatOpenAI(model="gpt-3.5-turbo")
mistral = Ollama(model="mistral")
ollama = Ollama(model="tinyllama")
```

```python
news_summary_prompt = ChatPromptTemplate.from_template("Summarize the following noisy news data and identify keywords of the news relevant to a Apple stock analyst. Provide a formatted answer as such Summary: ..., Keywords: ... News: {news}")
news_summary_chain = news_summary_prompt | gpt | StrOutputParser()
gpt_results = news_summary_chain.invoke({"news": sampled_news.iloc[-1]["fetched_body"]})
print(gpt_results)
```

```python
from langchain.chains import (
    StuffDocumentsChain,
    LLMChain,
    ReduceDocumentsChain,
    MapReduceDocumentsChain,
)
from langchain_core.prompts import PromptTemplate
from langchain_community.llms import OpenAI

# This controls how each document will be formatted. Specifically,
# it will be passed to `format_document` - see that function for more
# details.
document_prompt = PromptTemplate(
    input_variables=["page_content"],
     template="{page_content}"
)
document_variable_name = "context"
llm = OpenAI()
# The prompt here should take as an input variable the
# `document_variable_name`
prompt = PromptTemplate.from_template(
    "Summarize this content: {context}"
)
llm_chain = LLMChain(llm=ollama, prompt=prompt)
# We now define how to combine these summaries
reduce_prompt = PromptTemplate.from_template(
    "Combine these summaries: {context}"
)
reduce_llm_chain = LLMChain(llm=llm, prompt=reduce_prompt)
combine_documents_chain = StuffDocumentsChain(
    llm_chain=reduce_llm_chain,
    document_prompt=document_prompt,
    document_variable_name=document_variable_name
)
reduce_documents_chain = ReduceDocumentsChain(
    combine_documents_chain=combine_documents_chain,
)
chain = MapReduceDocumentsChain(
    llm_chain=llm_chain,
    reduce_documents_chain=reduce_documents_chain,
)

class Doc:
    page_content: str
    def __init__ (self, page_content):
        self.page_content = page_content
        self.metadata = []

chain.invoke({"input_documents": [Doc(sampled_news.iloc[-1]["fetched_body"])]})
```

```python
reduce_prompt = PromptTemplate.from_template(
    "Please summarize the following noisy but possible news data extracted from web page HTML, and extract keywords of the news. The news text can be very noisy due to it is HTML extraction. Give formatted answer such as Summary: ..., Keywords: ... The news is supposed to be for APPL stock. You may put ’N/A’ if the noisy text does not have relevant information to extract. News: {context}"
)
reduce_llm_chain = LLMChain(llm=ollama, prompt=reduce_prompt)
combine_documents_chain = StuffDocumentsChain(
    llm_chain=reduce_llm_chain,
    document_prompt=document_prompt,
    document_variable_name=document_variable_name
)
reduce_documents_chain = ReduceDocumentsChain(
    combine_documents_chain=combine_documents_chain,
)
```

```python
from langchain.chains.combine_documents.refine import RefineDocumentsChain
from langchain_core.prompts import PromptTemplate
from langchain.chains import (
    LLMChain,
)
document_prompt = PromptTemplate(
    input_variables=["page_content"],
     template="{page_content}"
)
document_variable_name = "context"
initial_prompt = PromptTemplate.from_template(
    "Summarize this content: {context}"
)
initial_chain = LLMChain(llm=ollama, prompt=initial_prompt)
initial_response_name = "prev_response"
reduce_prompt = PromptTemplate.from_template(
    "Here's your first summary: {prev_response}. "
    "Now add to it based on the following context: {context}"
)
reduce_llm_chain = LLMChain(llm=ollama, prompt=reduce_prompt)

chain = RefineDocumentsChain(
    initial_llm_chain=initial_chain,
    refine_llm_chain=reduce_llm_chain,
    document_prompt=document_prompt,
    document_variable_name=document_variable_name,
    initial_response_name=initial_response_name,
)
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator=".",
    chunk_size=200,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)
```

```python
text_splitter.create_documents([sampled_news.iloc[-1]["fetched_body"],])
```

```python
out = chain.invoke(text_splitter.create_documents([sampled_news.iloc[-1]["fetched_body"],]))
out
```
